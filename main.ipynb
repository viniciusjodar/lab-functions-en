## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def st_dev(list_sd):
    total = 0
    count = 0
    sq_div_sum = 0

    for n in list_sd:
        total += n
        count += 1

    mean = total / count

    for n in list_sd:
        div = n - mean
        sq_div_sum += div ** 2

    variance = sq_div_sum / (count - 1)
    return variance ** 0.5
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
    def splitt():
        words = ""
        word_list = []
        for char in string:
            if char == ",":
                word_list.append(words)
                words = ""
            else:
                words += char
        word_list.append(words)  # adiciona a Ãºltima palavra
        return word_list

    def joinn(words_to_join):
        if not words_to_join:
            return ""
        new_string = words_to_join[0]
        for i in range(1, len(words_to_join)):
            new_string += f",{words_to_join[i]}"
        return new_string

    list_words = splitt()
    list_words.sort()
    final_string = joinn(list_words)
    return final_string
```

```python
# This will test your function 
test_alpha(sort_alpha)
```
